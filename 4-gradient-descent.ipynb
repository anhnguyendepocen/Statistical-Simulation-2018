```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

## using gradient descent

```python
neta = 0.03
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
z = (x*sp.sin(20*y) + y*sp.sin(20*x)   )**2 * sp.cosh(sp.sin(10*x)*x) + (x*sp.cos(10*y) - y*sp.sin(10*x)   )**2  * sp.cosh(sp.cos(20*y)*y)
```

```python
zprimex = z.diff(x)
zprimey = z.diff(y)
```

```python
xi, yi = 0.5, 0.5
```

```python
for i in range(1000):
    xi = xi - neta*zprimex.evalf(subs = {x: xi, y:yi})
    yi = yi - neta*zprimey.evalf(subs = {x: xi, y:yi})
    
    if i%1 == 0:
        print(xi, yi)
```

# using simulated annealing

```python
thetals = [ [0.5, 0.5]]
T = 1
```

```python
def h(a):
    x = a[0]
    y = a[1]
    flag =    (x*np.sin(20*y) + y*np.sin(20*x)   )**2 * np.cosh(np.sin(10*x)*x) + (x*np.cos(10*y) - y*np.sin(10*x)   )**2  * np.cosh(np.cos(20*y)*y)
    return 1.0/flag
```

```python
def box_muller():
    a = np.zeros(2)
    u1 = np.random.uniform(size=1)
    u2 = np.random.uniform(size=1)
    a[0] = np.sqrt(-2*np.log(u1))* np.cos(2*np.pi*u2)
    a[1] = np.sqrt(-2*np.log(u1))* np.sin(2*np.pi*u2)
    return a
```

```python
for theta in thetals:
    for i in range(2000):
        old = theta
        G = theta + box_muller()
        u = np.random.uniform(size = 1)
        delh = h(G) - h(theta)
        if u>0 and u<min(1, np.exp(delh/T)):
            theta = G

    print(theta)
```
